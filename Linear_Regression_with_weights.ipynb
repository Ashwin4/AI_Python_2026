```python
#Linear regression with Gradient Descent
import numpy as np
x=np.array([[2,4],[4,7],[6,8],[8,9],[10,10]])
```

```python
x.shape
```

```python
x
```

```python
x.T
```

```python
samples,features=x.shape
```

```python
samples
```

```python
features
```

```python
y=np.array([50,60,70,80,96])
```

```python
np.zeros((3,2))
```

```python
w=np.zeros(features)#[w1,w2]
b=0
```

```python
w
```

```python
b
```

```python
epochs=100#going through the dataset fully one time here we taken 100 that means 100 times going thourgh the dataset
learning_rate=0.01
```

```python
y_pred=np.dot(x,w)+b#y=b+w1x1+w2x2
```

```python
y_pred
```

```python
y
```

```python
error=y-y_pred
```

```python
error.shape
```

```python
error
```

MAtrix multiplication rule

(3,4)(4,3) (2 5)(5,)

()

```python
y
```

```python
# This is without learning rate
for epoch in range(epochs):

    #Prediction
    y_pred=np.dot(x,w)+b#y=b+w1x1+w2x2

    error=y_pred-y

    #gradients
    dw=(1/samples)*np.dot(x.T,error)
    #if i slighlty change the w then how much loss change?
    #differentiaon(derivative) of weight wrt loss--
    #dl/dw=1/n(X.T)(y_predicted-y_actual)
    
    db=(1/samples)*np.sum(error)
    #derivative of loss wr.to bias
    #dl/db=1/n * total sum

    #update my parameters
    #learning rate is a parameter that controls how much the model updates itw weight dduring each step of training

    w=w-(dw)############without learning rate#############
    b=b-(db)

    print("epoch",epoch)
    print("weights and bias ",w,b)
    print("y_pred",y_pred)
    print()
```

```python
w=np.zeros(features)#[w1,w2]
b=0
```

```python
w
```

```python
b
```

```python
# This iw with a learning rare
for epoch in range(epochs):

    #Prediction
    y_pred=np.dot(x,w)+b#y=b+w1x1+w2x2

    error=y_pred-y

    #gradients
    dw=(1/samples)*np.dot(x.T,error)
    #if i slighlty change the w then how much loss change?
    #differentiaon(derivative) of weight wrt loss--
    #dl/dw=1/n(X.T)(y_predicted-y_actual)
    
    db=(1/samples)*np.sum(error)
    #derivative of loss wr.to bias
    #dl/db=1/n * total sum

    #update my parameters
    #learning rate is a parameter that controls how much the model updates itw weight dduring each step of training

    w=w-(learning_rate*dw)############### with learning rate #############
    b=b-(learning_rate*db)

    print("epoch",epoch)
    print("weights and bias ",w,b)
    print("y_pred",y_pred)
    print()
```
